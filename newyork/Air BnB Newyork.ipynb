```python
import pandas as pd
```

```python
import numpy as np
```

```python
import seaborn as sn
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Airbnb_NYC_2019.csv")
df
```

```python
airbnb_manhattan_df = df[df["neighbourhood_group"] == "Manhattan"]
airbnb_manhattan_df
```

```python
row, col = airbnb_manhattan_df.shape
print(row)
```

```python
print(col)
```

```python
plt.hist(airbnb_manhattan_df["price"], bins = np.arange(0, 1100, 40))
plt.title("Distribution of Airbnb listing Price in Manhattan for 2019")
plt.xlabel("Price (in US Dollars)")
plt.ylabel("Count")
plt.show()
```

Based on the histogram we can see that in 2019, Airbnb listings had the bulk of their price range between 50 to 500 US Dollars.

```python
plt.scatter(x = airbnb_manhattan_df["price"], y = airbnb_manhattan_df["number_of_reviews"], s = 5)
plt.xlim(0, 1100)
plt.title("Scatter plot of Number of Reviews vs Price of Airbnb listings in Manhattan")
plt.xlabel("Price (in US Dollars)")  
plt.ylabel("Numbber of Reviews")
plt.show()
```

Based on the analysis of scatter plot, it is visible that in 2019, Airbnb listings with lower prices that is price range between 50 to 500 US dollars gathered more reviews than those with high price. The more the price the lesser the reviews.

```python
value_counts_list =[]
value_counts = airbnb_manhattan_df['neighbourhood'].value_counts().nlargest(10)
value_counts_list.append(value_counts)
print (value_counts_list)
```

```python
value_counts = airbnb_manhattan_df['neighbourhood'].value_counts().nlargest(10)
df_value_counts = pd.DataFrame(value_counts)
df_value_counts = df_value_counts.reset_index()
df_value_counts.columns = ['neighbourhood', 'counts_top_manhattan']
df_value_counts
```

```python
row_count = df_value_counts.shape[0]
row_count
```

```python
col_count = df_value_counts.shape[1]
col_count
```

```python
df_all_cols = pd.merge(airbnb_manhattan_df, df_value_counts, how='inner')
```

```python
sn.countplot(x = "neighbourhood", data = df_all_cols)
plt.xticks(rotation=90, horizontalalignment='right')
plt.title("Airbnb Listing For Top Ten Neighburhood in Manhattan")
plt.show()
```

Based on the Countplot chart analysis, In 2019, the top 10 listings in the Manhattan district consisted mainly of the following neighourhoods Harlem, Upper West Side, Hells Kitchen, East Village, Upper East Side, Midtown, East Harlem, Chelsea, Lower East Side and Washington Heighs.
